```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
canvas_width = 101
canvas_height = 101
blank_character = " "
filled_character = "#"
null_character = "/"
starting_cell = int(canvas_width / 2)
canvas = []
```

```python
canvas = []
for row_number in range(canvas_height):
    
    row = []

    for cell_number in range(canvas_width):

        if (row_number == 0):
            if (cell_number == starting_cell):
                row.append(filled_character)
            else:
                row.append(blank_character)
        else:
            if (all_rules(canvas, row_number, cell_number)):
                row.append(filled_character)
            else:
                row.append(blank_character) 

    canvas.append(row)     
```

```python
data = np.array(canvas)

x = data[:, 0]
y = data[:, 1]
```

```python
# Create a scatter plot
plt.figure(figsize=(101, 101))
plt.scatter(x, y, color='blue', marker='o')

# Customize labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('2D Scatter Plot')

# Show grid
plt.grid(True)

# Display the plot
plt.show()
```

```python
def all_rules(canvas, row_number, cell_number):

    if (rule_1(canvas, row_number, cell_number)):
        return False
    
    if (rule_2(canvas, row_number, cell_number)):
        return True
    
    if (rule_3(canvas, row_number, cell_number)):
        return False
    
    if (rule_4(canvas, row_number, cell_number)):
        return True
    
    if (rule_5(canvas, row_number, cell_number)):
        return True

    if (rule_6(canvas, row_number, cell_number)):
        return False
    
    if (rule_7(canvas, row_number, cell_number)):
        return False
    
    if (rule_8(canvas, row_number, cell_number)):
        return True
```

```python
def rule_1(canvas, row_number, cell_number):

    first_case = False
    second_case = False
    third_case = False

    try:
        top_left_cell = canvas[row_number - 1][cell_number - 1]
    except IndexError:
        top_left_cell = null_character

    top_middle_cell = canvas[row_number - 1][cell_number]

    try:
        top_right_cell = canvas[row_number - 1][cell_number + 1]
    except IndexError:
        top_right_cell = null_character

        
    if (top_left_cell == filled_character):
        third_case = True
        
    if (top_middle_cell == filled_character):
        first_case = True

    if (top_right_cell == filled_character):
        second_case = True

    if first_case and second_case and third_case:
        return True
    else:
        return False
```
